```python
df.drop(columns=['sg', 'su', 'rbc', 'pcc', 'ba', 'bu',
        'sod', 'pot',  'pcv', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane'],inplace=True)
```

```python
num_col=[col for col in df.columns if df[col].dtype!="object"]

for col in num_col:
    df[col]=df[col].fillna(df[col].median())
```

```python
cat_col=[col for col in df.columns if df[col].dtype=="object"]

for col in cat_col:
    print(f" {col} most frequent value is {df[col].mode()}\n")
```

```python
x = df.drop(columns=['classification'])
y = df['classification']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print("Accuracy: ", accuracy*100)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```

```python
from xgboost import XGBClassifier

model_2 = XGBClassifier()

model_2.fit(X_train, y_train)
accuracy = model_2.score(X_test, y_test)
print("Accuracy: ", accuracy*100)
```

```python
y_pred = model_2.predict(X_test)
from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```

```python
from sklearn import svm

model_3 = svm.SVC(kernel='linear')
#training the support vector Machine Classifier
model_3.fit(X_train,y_train)
accuracy = model_3.score(X_test, y_test)
print("Accuracy: ", accuracy*100)
```

```python
y_pred = model_3.predict(X_test)
from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```
